```python
# 결측치 :NaN의미 => 결측치 제거 함수를 이용
# 이상치 : NaN이 아님 => 결측치로 바꾸고 => 결측치 제거 함수로 제거
```

```python
# 1.
import numpy as np
df=df.replace("?",np.NaN)
df=df.dropna()
```

```python
# 2.
df[df['horsepower'] != "?"]
```

```python
# 3.
# horsepower 컬럼의 결측치만 제거
df.dropna(subset = ['horsepower'], axis=0, inplace=True)

# inplace=True : 삭제 후 메모리에 반영 시키기
# df 데이터프레임에는 ?가 없어짐
```

```python
# horsepower 컬럼의 타입을 float 타입으로 변경하기
df['horsepower']=df['horsepower'].astype(float)
```
