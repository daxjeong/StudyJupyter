```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./file_new/auto-mpg_v2.csv")
df
```

```python
# 사용할 변수들 : mpg, cylinders, horsepower, weight
ndf = df[['mpg','cylinders','horsepower','weight']]
ndf
```

```python
# 선형회귀분석 모듈
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# 다항식으로 변환하는 모듈(2차원으로 변환해요 사용)
from sklearn.preprocessing import PolynomialFeatures
```

```python
# 예측모델
x = np.array(ndf[['weight']])
y = ndf["mpg"]
```

```python
# train 80%, test 30%로 구분하여 조회
x_train, x_test, y_train, y_test = train_test_split(x,
                                                    y,
                                                    test_size = 0.3,
                                                    random_state = 10)
```

```python
poly = PolynomialFeatures(degree = 2)
```

```python
x_train_poly = poly.fit_transform(x_train)
```

```python
x_train_poly
```

```python
pr = LinearRegression()
```

```python
# 예측
pr.fit(x_train_poly, y_train)
```

```python
# 검증
x_test_poly = poly.fit_transform(x_test)
```

```python
# 학습시킨 모델에 검증 데이터를 넣어서 결정계수 확인하기
pr.score(x_test_poly, y_test)
```

```python
y_hat_test = pr.predict(x_test_poly)
```

```python
y_hat_test
```

```python
fig = plt.figure(figsize = (10, 5))
ax = fig.add_subplot(1, 1, 1)
ax.plot(x_train, y_train   , "go", label = "Orginal Data"  )
ax.plot(x_test , y_hat_test, "r+", label = "Predicted Data")

ax.legend(loc = "best")

plt.xlabel("weight")
plt.ylabel("mpg")

plt.show()
plt.close()
```

```python
x_poly = poly.fit_transform(x) # 2차원으로 바꿔줌
y_hat = pr.predict(x_poly)

fig = plt.figure(figsize=(10,5))

ax1 = sns.kdeplot(y, label ="y")
ax2 = sns.kdeplot(y_hat, label ="y_hat",ax=ax1)

plt.legend()
plt.show()
```
