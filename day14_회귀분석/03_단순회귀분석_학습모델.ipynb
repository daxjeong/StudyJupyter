```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./file_new/auto-mpg_v2.csv")
```

```python
df
```

```python
# 지도학습 모델 : 단순회귀분석(선형회귀모델, 선형회귀예측모델...)
# - 종속변수가 있어요
# - 직선의 방정식을 사용 : y = ax + b
# - y : 종속변수(정답)
# - x : 독립변수
# - a : 기울기
# - b : y절편

# - 학습 : fit
```

```python
# 사용할 변수들 : mpg, cylinders, horsepower, weight
ndf = df[['mpg','cylinders','horsepower','weight']]
ndf
```

```python
# 산점도 그래프 그리기
# 종속변수:mpg, 독립변수:weight
plt.figure(figsize=(10,5))
plt.scatter(ndf['weight'],ndf['mpg'],color="red", s=10)
plt.show()
```

```python
ndf.plot(kind='scatter',
         x = 'weight',
         y = 'mpg',
         c = 'deeppink',
         s = 10,
         figsize=(10,5))
plt.show()
plt.close()
```

```python
# 해석
# 중량이 커질수록 연비는 낮아지는 음의 상관관계
# 중량은 연비에 영향을 미친다
```

```python
# 회귀선(추세선) 그리기
ndf.plot(kind='scatter',
         x = 'weight',
         y = 'mpg',
         c = 'deeppink',
         s = 10,
         figsize=(10,5))
sns.regplot(x='weight', y='mpg', ci=None, data=ndf)
plt.show()
plt.close()
```

```python
plt.subplot(1,2,1)
plt.subplot(1,2,2)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(ndf['weight'],ndf['mpg'],color="red", s=10)

ndf.plot(kind='scatter',
         x = 'weight',
         y = 'mpg',
         c = 'deeppink',
         s = 10,
         figsize=(6,4))
sns.regplot(x='weight', y='mpg', ci=None, data=ndf)
plt.show()
plt.close()
```

```python
# 회귀선(추세선) 그리기
fig = plt.figure(figsize=(10,5))

# 첫번째 그래프 공간 만들기(1행 2열의 1열)
ax1 = fig.add_subplot(1,2,1)

# 두번째 그래프 공간 만들기(1행 2열의 2열)
ax2 = fig.add_subplot(1,2,2)

# 첫번째 그래프
sns.regplot(x = "weight",
            y = "mpg",
            data = ndf,
            ax = ax1,
            color = "purple",
            fit_reg = True)

# 두번째 그래프
sns.regplot(x = "weight",
            y = "mpg",
            data = ndf,
            ax = ax2,
            color = "teal",
            fit_reg = False)

plt.show()
plt.close()
```
