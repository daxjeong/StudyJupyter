# 3day 빈도 및 교차 데이터 만들기

### tips 데이터셋 읽어들이기

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pylab as plt
import seaborn as sns
import numpy as np
```

```python
# tips 데이터셋 읽어들이기
tips = sns.load_dataset("tips")
tips
```

### 빈도분석

```python
tips[['sex']] # 총 244건, range(0,243)
```

```python
# 성별 빈도분석
tips['sex'].value_counts()
```

```python
# 월별/성별 교차분석
pd.crosstab(tips["sex"], tips["day"])
```

```python
# 성별/월별 교차분석 총괄표
pd.crosstab(tips["sex"], tips["day"], margins = True)
```

```python
# 빈도 비율 확인
pd.crosstab(tips["sex"], 
            tips["day"]).apply(
                lambda r: round(r/len(tips)*100,2),
                axis = 1)
```
