# 컬럼명 한글명으로 변환
### -샘플로 trfcard(0) 폴더 1개만 우선 변환

```python
import pandas as pd
```

```python
trfcard_0 = pd.read_csv("./01_data_원본/포항시_버스_사용자_교통카드_사용_내역/trfcard (0)/trfcard.csv")
trfcard_0.head()
```

```python
trfcard_col = pd.read_excel("./01_data_원본/포항시_버스_사용자_교통카드_사용_내역/trfcard (0)/trfcard_columns.xlsx",
              skiprows =2,
              usecols="B,C")

trfcard_col
```

```python
trfcard_col_list=trfcard_col.to_dict('records')
trfcard_col_list
```

```python
new_dict={}
for i in range(0, len(trfcard_col_list)):
    new_dict[trfcard_col_list[i]['컬럼명 (영문)']] = trfcard_col_list[i]['컬럼명 (한글)']
print(new_dict)
```

```python
trfcard_0_kor = trfcard_0.rename(columns = new_dict)
trfcard_0_kor.head(1)
```

```python
# trfcard(0~99) 통합하기
df_all=pd.DataFrame()

df_dict={}
df_col_list={}
for i in range(0,100):
    df_dict[i]=pd.read_csv("./01_data_원본/포항시_버스_사용자_교통카드_사용_내역/trfcard ({})/trfcard.csv".format(i))
    df_col_list[i] = pd.read_excel("./01_data_원본/포항시_버스_사용자_교통카드_사용_내역/trfcard ({})/trfcard_columns.xlsx".format(i),
              skiprows =2,
              usecols="B,C")

    new_col_list = {}
    
    for j in range(0, len(df_col_list[i])):
        new_col_list[df_col_list[i].to_dict('records')[j]['컬럼명 (영문)']] = df_col_list[i].to_dict('records')[j]['컬럼명 (한글)']
    
    df_dict[i]=df_dict[i].rename(columns = new_col_list)
    

    df_all = df_all.append(df_dict[i], ignore_index = True)
```
