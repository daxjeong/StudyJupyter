```python
def chart(x):
    df_filtering=df[df['국적']== x]
    plt.figure(figsize = (10, 2))

    plt.plot(df_filtering['기준년월'], df_filtering['관광'])
    plt.title("{}에 대한 관광객 추이 분석".format(x))
    plt.xlabel("기준년월")
    plt.ylabel("관광객수")
    plt.xticks(['2010-01','2011-01','2012-01','2013-01','2014-01','2015-01','2016-01','2017-01','2018-01','2019-01'])
    return plt.show()
```

```python
cntry_list = ['중국', '일본', '대만', '미국', '홍콩']
for cntry in cntry_list:
    chart(cntry)
```

```python
# 히트맵 : 전체 시계열 데이터에 대한 이슈를 한 눈에 보기
# 변수 : 3개가 필요
```

```python
# 년도와 월만 조회
# 기준년월에서 년도만 추출하기
df['기준년월'].str[:4]
df['년도']=df['기준년월'].str.slice(0,4)
```

```python
# 기준년월에서 월만 추출하기
df['기준년월'].str[5:]
df['월']=df['기준년월'].str.slice(5,7)
```

```python
df.head()
```

```python
df_filter = df[(df['국적']=='중국')]
df_filter
```

```python
# 히트맵 그리는 순서
# 1. 피봇으로 데이터 집계
# 2. 피봇 결과를 히트맵에 적용
```

```python
# 1. 피봇 데이터 만들기
df_pivot = df_filter.pivot_table(values = '관광',
                      index = '년도',
                      columns = '월')
df_pivot
```

```python
# 2. 히트맵 그리기
# 히트맵은 시본 라이브러리를 함께 사용해야 함
import seaborn as sns
```

```python
# 그래프 크기 설정
plt.figure(figsize = (13, 10))

# 피봇 테이블 넣어서 그래프 그리기
sns.heatmap(df_pivot, annot = True, fmt = ".0f", cmap = "rocket_r")

# 타이틀 넣기
plt.title("년/월별 [중국] 관광객 히트맵 시각화분석")

# 보여주기
plt.show()
```

```python
# 중국, 일본, 대만, 미국, 홍콩 히트맵 그리기
cntry_list = ['중국', '일본', '대만', '미국', '홍콩']
for cntry in cntry_list:
    df_filter = df[(df['국적']== cntry)]
    df_pivot =  df_filter.pivot_table(values = '관광',
                          index = '년도',
                          columns = '월')

    plt.figure(figsize = (13,10))
    sns.heatmap(df_pivot, annot = True, fmt = ".0f", cmap = "rocket_r")
    plt.title("년/월별 {} 관광객 히트맵 시각화분석".format(cntry))
    plt.show()
```
