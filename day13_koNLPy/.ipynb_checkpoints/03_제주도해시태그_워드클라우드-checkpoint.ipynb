```python
import pandas as pd
```

```python
raw_total = pd.read_excel("./files/3_1_crawling_raw.xlsx")
raw_total
```

```python
raw_total['tags']
```

```python
tags_total = []
# 해시태그 단어 하나하나를 모두 추출하여 하나의 리스트로 만들기
for tags in raw_total['tags'] :
    for tag in tags[2:-2].split("', '") : 
        tags_total.append(tag.replace('\'', ''))
tags_total
```

```python
from collections import Counter
```

```python
tag_counter = Counter(tags_total)
tag_counter
```

```python
# 가장 많은 count 조회
tag_counter.most_common(50)
```

```python
# 키워드 정제하기
# 정제할 키워드에 대한 리스트 만들기
STOPWORDS = ["#일상", "#선팔", "#제주도", "#반영구",
             "#제주자연눈썹", "#서귀포눈썹문신", "#제주눈썹문신", "#소통", "#맞팔"]
```

```python
tag_total_selected = []

for tag in tags_total:
    if tag not in STOPWORDS:
        tag_total_selected.append(tag)
        
tag_total_selected
```

```python
tag_total_selected = Counter(tag_total_selected)
tag_total_selected
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
from matplotlib import font_manager, rc
import platform
```

```python
path = "c:/Windows/Fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname=path).get_name()
rc("font", family=font_name)
```

```python
# 막대그래프(빈도그래프)
tag_counts_df = pd.DataFrame(tag_total_selected.most_common(30))
tag_counts_df
```

```python
tag_counts_df.columns = ["tags","counts"]
tag_counts_df
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x="counts",
            y="tags",
            data=tag_counts_df)
```
