# 2. 재구조화

```python
# cut() : 동일한 길이(값의 범위)로 3구간으로 나눈다
df1 = pd.cut(df['math'],3)
df1

# 결과는 수학점수가 어느 구간에 속하는지 보여줌
```

```python
df2 = df['math'].groupby(df1)
```

```python
df2.head(1)
```

```python
# df2에 대한 기초통계 : agg() 함수 사용
df3 = df2.agg(['count','mean','std','min','max'])
df3
```

```python
# qcut() : 동일개수로 나누기

pd.qcut(df['math'], 3)
```

```python
# labels=False : 범주의 번호 표시하기
df4 = pd.qcut(df['math'], 3, labels=False)
df4
```

```python
import numpy as np
```

```python
df5 = pd.qcut(df['math'],3,labels = np.arange(3, 0 , -1))
df5
```

```python
# range()
# 첫번째값 : 시작값
# 두번째값 : 끝 값 -1 
# 세번째값 : 증가값(+) or 감소값(-)
for i in range(0, 3, 1):
    print(i)
    
print("--------------")

# 증가값을 감소 시킬때는 두번째 값은 +1 
# 두번째값 : 0보다 하나 클 때까지
for j in range(3, 0, -1):
    print(j)
```

```python
# np.arange()
# 첫번째값 : 시작값
# 두번째값 : 끝 값 -1 
# 세번째값 : 증가값(+) or 감소값(-)
for i in np.arange(0, 3, 1):
    print(i)
    
print("--------------")

for j in np.arange(3, 0, -1):
    print(j)
```

```python
df6 = df['math'].groupby(df5)
```

```python
df6.head()
# 수학점수에 대해 3,2,1로 그룹
```

```python
# 기초통계 조회
# count, mean, std, min, max
df7 = df6.agg(('count','mean', 'std', 'min', 'max'))
df7

# NAN : 한 명밖에 없어서 중앙값이 없음
```
