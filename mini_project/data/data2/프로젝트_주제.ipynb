# 시청률은 가구시청률 이용

```python
rate = pd.read_csv("./종합반응DB시청률_total.csv", encoding='cp949')
rate
```

```python
rate[['TVVOD시청자수(K)', 'PC실시간(K)', 'PCVOD(K)', 'Mobile실시간(K)', 'MobileVOD(K)' ]]
```

```python
drama_rate = rate[['프로그램명','채널','요일','게시글수','댓글수','동영상조회(K)','뉴스','동영상수','가구시청률(%)']]
```

```python
# 가구시청률이 0인 row 제거
drama_rate = drama_rate[drama_rate['가구시청률(%)'] != 0]
drama_rate
```

```python
df_group= drama_rate.groupby(['프로그램명'])
df_group.head()
```

```python
drama_rate['가구평균시청률(%)'] = df_group[['가구시청률(%)']].transform('mean')
drama_rate.sort_values(by='가구평균시청률(%)',ascending=False)
```
