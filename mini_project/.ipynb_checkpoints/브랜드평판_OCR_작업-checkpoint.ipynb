```python
import pytesseract
from PIL import Image

img_path = "78111_183242_5724.jpg"
image_pil = Image.open(img_path)
img_pytesseract_en = pytesseract.image_to_string(image_pil, lang="kor")
```

```python
aa = img_pytesseract_en.split()[12:]
n=7
result1 = [aa[i * n:(i + 1) * n] for i in range((len(aa) - 1 + n) // n )] 

import pandas as pd
a = pd.DataFrame(result1)
a
```

```python
# 인덱스 1위부터 10위
finish = a[[0,5]][0:10]
finish.index = finish.index+1

# 열 이름 변경
finish.columns = ['드라마배우','브랜드평판지수']
finish
```

```python
img_path = "bp201702.png"
image_pil = Image.open(img_path)
img_pytesseract_en = pytesseract.image_to_string(image_pil, lang="kor")

img_pytesseract_en
```

```python
aa = img_pytesseract_en.split()[3:]
aa



n=3
result1 = [aa[i * n:(i + 1) * n] for i in range((len(aa) - 1 + n) // n )] 

import pandas as pd
a = pd.DataFrame(result1)
a

# 인덱스 1위부터 10위
finish = a[[0,1]]
finish.index = finish.index+1

# 열 이름 변경
finish.columns = ['드라마배우','브랜드평판지수']
finish

finish.to_csv("./data/brand_score/bp201702.csv")
```
