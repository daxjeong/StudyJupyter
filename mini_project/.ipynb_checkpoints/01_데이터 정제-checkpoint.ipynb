# 컬럼명 영문명을 한글명으로 변환

```python
# 메타정의서 불러들이기(영문명과 한글명이 있는 컬럼만 불러들이기)
index = pd.read_excel("./data/K-드라마 프로그램 출연진정보 채_컬럼정의서.xls", usecols="B,C")
```

```python
# 딕셔너리로 바꾸기 - 1단게(완전한 딕셔너리는 아님)
# 리스트 형태로 변환됨 -> 내용은 딕셔너리
# records : 데이터프레임의 행들을 의미 -> 각 행들을 딕셔너리로 전환해서 가지고오기

index_kor = index.to_dict("records")
index_kor

# to_dict("records") : 각 행을 사전형으로 변환
```

```python
# 완전한 딕셔너리 만들기 - 2단계(key와 value를 직접 만들기)

new_dict={}
for i in range(0, len(index_kor)):
    new_dict[index_kor[i]['컬럼영문명']] = index_kor[i]['컬럼한글명']
print(new_dict)
```

```python
# # 컬럼명 변경하기(rename)
df_all = {}
for i in index_kor:
    df_all[i['컬럼영문명']] = i['컬럼한글명']
df_new = df.rename(columns=df_all)
```

```python
df_new
```

```python
# 프로그램 시작시간이 21~22시
# 채널 : KBS2,MBC,SBS,OCN,JTBC,TVN
# 미니드라마 시청률 분석
```

```python
# 프로그램 시작시간 앞 두자리 추출하기 위해서
# 문자타입으로 변환해주기

# astype : 열의 요소의 dtype을 변경하는함수 
df_new['프로그램시작시간'].astype(str)
```

```python
a = df_new['프로그램시작시간'].astype(str)
a
```

```python
# 모든 프로그램시작시간을 동일한 6자리로 만들어 주기위해 0으로 채워줌
# zfill : 파이썬 문자열 앞 0으로 채우기
a[1] = a[1].zfill(6)
a
```

```python
for i in range(0, len(a)):
    a[i] = a[i].zfill(6)
print(a)

```

```python
df_new['프로그램시작시간']=a
df_new
```

```python
# 프로그램 시작시간 앞의 2자리만 추출
# 저녁 9시 ~ 10시 드라마만 분석(21~22)

df_new['프로그램시작시간']=df_new['프로그램시작시간'].str[0:2]
df_new.head(1)
```

```python
# 분석에 사용할 컬럼만 필터링

drama = df_new[['방송일자','채널명','프로그램시작시간','프로그램종료시간','프로그램명','방송회차명','프로그램장르소분류명','출연진내용','총시청률']]
drama.head()
```

```python
# 프로그램장르가 미니시리즈인 드라마만 필터링

drama_mini=drama[drama['프로그램장르소분류명']=='미니시리즈']
drama_mini
```

```python
# 21시 ~ 22시에 방영되는 드라마 필터링

drama_mini_2122 = drama_mini[(drama_mini['프로그램시작시간'] == '21')|(drama_mini['프로그램시작시간'] == '22')]
drama_mini_2122
```

```python
drama_mini_2122['프로그램명'].unique()
```

```python
# 분석 할 6가지 채널만 필터링

drama_202001=drama_mini_2122[(drama_mini_2122['채널명']=='KBS2')|(drama_mini_2122['채널명']=='MBC')|(drama_mini_2122['채널명']=='SBS')|(drama_mini_2122['채널명']=='JTBC')|\
               (drama_mini_2122['채널명']=='tvN') | (drama_mini_2122['채널명']=='OCN')]
drama_202001=drama_202001.reset_index(drop=True)
```
