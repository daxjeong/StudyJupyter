```python
import pandas as pd
```

```python
# 여기에 통합할 프로그램 작성
df_drama_kor_tot = pd.DataFrame()

year_month=[202001,202002,202003,202004,202005,202006,202007,202008,202009,202010,202011,202012,202101,202104,202105,202107,202108,202109,202110,202111,202112]
for ym in year_month:
    # 1. csv, excel파일 불러들이기
    file_path1="./data/TN_K_DRAMA_CST_AUDERT_INFO_{}.csv".format(ym)
    file_path2="./data/K-드라마 프로그램 출연진정보 채_컬럼정의서.xls"
    
    df_drama = pd.read_csv(file_path1)
    df_drama_col = pd.read_excel(file_path2, usecols="B:C")
    
    # 2. 컬럼명 한글로 바꾸기
    df_drama_col_dict = df_drama_col.to_dict("records")
    
    df_drama_new_col_dict = {}
    
    for i in range(0, len(df_drama_col_dict)):
        df_drama_new_col_dict[df_drama_col_dict[i]['컬럼영문명']] = df_drama_col_dict[i]['컬럼한글명']
    
    df_drama_kor = df_drama.rename(columns=df_drama_new_col_dict)
    
    df_drama_kor_tot = df_drama_kor_tot.append(df_drama_kor, ignore_index = True)
    
    
```

```python
# 총시청률 구하기
df_drama_kor_tot['총시청률'] = (df_drama_kor_tot['남자4_9세시청율']+df_drama_kor_tot['남자10대시청율']+df_drama_kor_tot['남자20대시청율']+df_drama_kor_tot['남자30대시청율']+df_drama_kor_tot['남자40대시청율']+df_drama_kor_tot['남자50대시청율']+df_drama_kor_tot['남자60대이상시청율'] + \
                df_drama_kor_tot['여자4_9세시청율']+df_drama_kor_tot['여자10대시청율']+df_drama_kor_tot['여자20대시청율']+df_drama_kor_tot['여자30대시청율']+df_drama_kor_tot['여자40대시청율']+df_drama_kor_tot['여자50대시청율']+df_drama_kor_tot['여자60대이상시청율'])/14
```

```python
# 컬럼명 정리
df_drama_kor_tot=df_drama_kor_tot[['방송일자','채널명','프로그램시작시간','프로그램종료시간','프로그램명','방송회차명','프로그램장르소분류명','출연진내용','총시청률']]
df_drama_kor_tot.head()
```

```python
# 시작시간 21시-22시 추출
a=df_drama_kor_tot['프로그램시작시간'].astype(str)

for i in range(0, len(a)):
    a[i] = a[i].zfill(6)
print(a)

df_drama_kor_tot['프로그램시작시간'] = a
df_drama_kor_tot['프로그램시작시간']=df_drama_kor_tot['프로그램시작시간'].str[0:2]
```

```python
df_drama_kor_tot = df_drama_kor_tot[(df_drama_kor_tot['프로그램시작시간'] == '21')|(df_drama_kor_tot['프로그램시작시간'] == '22')]
df_drama_kor_tot.head()
```

```python
# 미니드라마
df_drama_kor_tot=df_drama_kor_tot[df_drama_kor_tot['프로그램장르소분류명']=='미니시리즈']
df_drama_kor_tot.head()
```

```python
# 채널
# KBS2, SBS, MBC, JTBC, TVN, OCN
df_drama_kor_tot=df_drama_kor_tot[(df_drama_kor_tot['채널명']=='KBS2')|(df_drama_kor_tot['채널명']=='MBC')|(df_drama_kor_tot['채널명']=='SBS')|(df_drama_kor_tot['채널명']=='JTBC')|\
               (df_drama_kor_tot['채널명']=='tvN') | (df_drama_kor_tot['채널명']=='OCN')]
df_drama_kor_all=df_drama_kor_tot.reset_index(drop=True)
```

```python
# 이상치 제거
df_drama_kor_all= df_drama_kor_all[~df_drama_kor_all['출연진내용'].str.contains("저자")] # 출연진 오류
df_drama_kor_all= df_drama_kor_all[~((df_drama_kor_all['프로그램명'].str.contains("어서와"))|(df_drama_kor_all['프로그램명'].str.contains("꼰대인턴방구석팬미팅")|\
                                                                                     (df_drama_kor_all['프로그램명'].str.contains("드라마맛집JTBC7월의레시피"))))] # 예능
df_drama_kor_all=df_drama_kor_all[~df_drama_kor_all['프로그램명'].str.contains("아이를찾습니다")] # 2부작이라서 제외




df_drama_kor_all= df_drama_kor_all[df_drama_kor_all['출연진내용'].str.contains(":", na = False )].reset_index(drop=True)
```

```python
df_drama_kor_all=df_drama_kor_all.reset_index(drop=True)
```

```python
# 주요 출연진 3명 추출
a=[]
for i in range(0,len(df_drama_kor_all)):
    actor=[]
    a.append(df_drama_kor_all['출연진내용'][i].split(',')[0:3])
    for j in range(0, 3):
        actor.append(a[i][j].split(":")[1])
        s = ",".join(actor)
        df_drama_kor_all['출연진내용'][i] = s
```
