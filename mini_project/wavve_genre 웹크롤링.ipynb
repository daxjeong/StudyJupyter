```python
#4 For 구문 적용

from selenium import webdriver

from bs4 import BeautifulSoup

import time

genrelist = []

for i in title:
    try:
        
        driver = webdriver.Chrome("C:/ChromeDriver_exe/chromedriver.exe")

        URL =  'https://www.wavve.com/search/search?category=program&searchWord={}'.format(i)

        driver.get(URL)

        html = driver.page_source

        #soup = BeautifulSoup(html, "html.parser")

        driver.find_element_by_xpath('//*[@id="contents"]/div[3]/div/div/div/div[2]/div[1]/a/div[2]/div/div/span').click()

        #URL2 = driver.current_url

        time.sleep(0.5)
        #driver.get(URL2)

        #html = driver.page_source

        soup = BeautifulSoup(html, "html.parser")

        driver.find_element_by_xpath('//*[@id="contents"]/section/div/div[2]/div[5]/button/span').click()

        time.sleep(0.5)

        td = driver.find_elements_by_css_selector('td')
        genre = td[1].text

        genre = genre.split(",")[1:]


        name = soup.select("div.list-view-detail>div")
        search = name[0].text

        if title == search:
            genrelist.append('없음')
        else:
            genrelist.append(genre)
                     
    except:
        genrelist.append('없는드라마')
```
